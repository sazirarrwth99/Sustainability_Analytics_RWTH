```python
import pandas as pd
import statsmodels.api as sm
from scipy.stats import pearsonr

def check_causality(X, Y):
    """
    Check for causality between two time series using three approaches:
    1. Granger causality test
    2. Pearson correlation
    3. Vector autoregression (VAR)
    
    Parameters:
    - X (array-like): Time series data for series X
    - Y (array-like): Time series data for series Y
    
    Returns:
    - result (dict): Dictionary with the test results for each approach
    """
    
    # Convert the time series to a pandas DataFrame
    df = pd.DataFrame({'X': X, 'Y': Y})
    
    result = {}
    
    # Approach 1: Granger causality test
    granger_test = sm.tsa.stattools.grangercausalitytests(df, maxlag=3, verbose=False)
    result['granger_test'] = granger_test
    
    # Approach 2: Pearson correlation
    corr, pval = pearsonr(X, Y)
    result['pearson_correlation'] = (corr, pval)
    
    # Approach 3: Vector autoregression (VAR)
    model = sm.tsa.VAR(df)
    results = model.fit(maxlags=3)
    causal_matrix = results.test_causality('X', 'Y')
    result['var_causality'] = causal_matrix
    
    return result
```

```python
import pandas as pd
import statsmodels.api as sm
from scipy.stats import pearsonr

def check_causality(X, Y, alpha=0.05):
    """
    Check for causality between two time series using three approaches:
    1. Granger causality test
    2. Pearson correlation
    3. Vector autoregression (VAR)
    
    Parameters:
    - X (array-like): Time series data for series X
    - Y (array-like): Time series data for series Y
    - alpha (float): Significance level for the statistical tests (default: 0.05)
    
    Returns:
    - result (tuple): Tuple with the test results for each approach
    """
    
    # Convert the time series to a pandas DataFrame
    df = pd.DataFrame({'X': X, 'Y': Y})
    
    result = []
    
    # Approach 1: Granger causality test
    granger_test = sm.tsa.stattools.grangercausalitytests(df, maxlag=3, verbose=False)
    pvalues = [t[1] for t in granger_test[3]['ssr_chi2test']]
    causality = any(pval < alpha for pval in pvalues)
    result.append(causality)
    
    # Approach 2: Pearson correlation
    corr, pval = pearsonr(X, Y)
    result.append(pval < alpha)
    
    # Approach 3: Vector autoregression (VAR)
    model = sm.tsa.VAR(df)
    results = model.fit(maxlags=3)
    causal_matrix = results.test_causality('X', 'Y')
    pvalues = causal_matrix['pvalue'].values
    causality = any(pval < alpha for pval in pvalues)
    result.append(causality)
    
    return tuple(result)
```
